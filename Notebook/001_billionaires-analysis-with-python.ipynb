# **Billionaires Analysis with Python**

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
```

**So let’s get started with the task of billionaires analysis by importing the necessary Python libraries and the dataset:**

```python
data = pd.read_csv("/Users/robinkraj/Desktop/3_prof_work1/2/56_Canada_DA/MachingLearningProjects/Data/Forbes Billionaires.csv")
data.head()
```

**Before we go ahead, let’s see whether or not this dataset contains missing values:**

```python
data.isnull().sum()
```

**Now let’s have a look at the top 10 billionaires according to their NetWorth:**

```python
df = data.sort_values(by=["Networth"],ascending = False).head(10)
plt.figure(figsize = (20,10))
sns.histplot(x="Name", hue ="Networth",data=df)
plt.show()
```

**Now let’s have a look at the top 5 domains with the most number of billionaires:**

```python
a = data['Source'].value_counts().head()
index = a.index
source = a.values
plt.figure(figsize = (5,5))
plt.pie(source, labels = index)
central_circle = plt.Circle((0,0),0.5,color = 'white')
# fig = plt.gcf()
plt.gca().add_artist(central_circle)
# plt.rc('font', size =12)
plt.title("top 5 domains to become billionare", fontsize=20)
plt.show()
```

```python
a
```

**Now let’s have a look at the top 5 industry with the most number of billionaires:**

```python
b = data['Industry'].value_counts().head()
index = b.index
industry = b.values
plt.figure(figsize = (5,5))
plt.pie(industry, labels = index)
central_circle = plt.Circle((0,0),0.5,color = 'white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
# plt.rc('font', size =12)
plt.title("top 5 industry to become billionare", fontsize=20)
plt.show()
```

```python
b
```

**Now let’s have a look at the top 5 Country with the most number of billionaires:**

```python
c = data['Country'].value_counts().head()
index = c.index
source = c.values
plt.figure(figsize = (5,5))
plt.pie(source, labels = index)
central_circle = plt.Circle((0,0),0.5,color = 'white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
# plt.rc('font', size =12)
plt.title("top 5 country to become billionare", fontsize=20)
plt.show()
```
